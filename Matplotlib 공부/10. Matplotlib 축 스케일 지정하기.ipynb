## X축 스케일 지정하기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100) # -10부터 10사이를 숫자 100개로 채움
y = x ** 3

plt.plot(x, y)
plt.xscale('symlog')

plt.show()
```

## Y축 스케일 지정하기

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('linear') #선형 스케일

plt.show()
```

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('log') #로그 스케일

plt.show()
```
