```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo') #bo = 'b'는 blue, 'o'= circle를 나타냅니다.
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

## 선/마커 동시에 나타내기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo-') #파란색 + 실선 + 마커
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo--') #파란색 + 점선 + 마커
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

## 선/마커 표시 형식

```python
plt.plot([4, 5, 6], marker="H")
plt.plot([3, 4, 5], marker="d")
plt.plot([2, 3, 4], marker="x")
plt.plot([1, 2, 3], marker=11)
plt.plot([0, 1, 2], marker='$Z$')
plt.show()
```
