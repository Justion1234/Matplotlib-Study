```python
import matplotlib.pyplot as plt

plt.plot((0, 2), (1, 3), (2, 5), (3, 10))
plt.show()
```

## x,y 값 입력하기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.show()
```

```python
plt.plot([0, 2], [1, 3], [2, 5], [3, 10])
plt.show()
```

## 테이블이 있는 데이터 사용하기

```python
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```
