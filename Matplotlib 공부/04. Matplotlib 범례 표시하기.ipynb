범례(Legend)는 그래프 데이터의 종류를 표시하기 위한 텍스트이다.
# 
matplotlib.pyplot 모듈의 legend() 함수를 사용해서 그래프에 범례를 표시할 수 있습니다.

## 기본 사용

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend() #이 코드를 통해서 범례를 그래프에 나타내줌

plt.show()
```

## 위치 지정하기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(1.0, 1.0))

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(0, 0))

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(0.5, 0.5))

plt.show()
```

xlabel(), ylabel() 함수의 labelpad 파라미터는 축 레이블의 여백(Padding)을 지정합니다.
legend() 함수의 loc 파라미터를 이용해서 범례가 표시될 위치를 설정할 수 있습니다.
loc 파라미터를 숫자 쌍 튜플로 지정하면, 해당하는 위치에 범례가 표시됩니다.
loc=(0, 0, 0, 0)은 데이터 영역의 왼쪽 아래, loc=(1.0, 1.0)은 데이터 영역의 오른쪽 위 위치입니다

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='lower right') #오른쪽 아래에 범례가 표시됩니다.

plt.show()
```

## 열 개수 지정하기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best')          # ncol = 1


plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demad (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

#loc='best'  -> 자동으로 계산하여 최적의 위치에 레전드를 위치시킨다.

plt.legend(loc='best', ncol=2)

plt.show()
```

## 폰트 크기 지정하기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best', ncol=2, fontsize=14)

plt.show()
```

## 범례 테두리 꾸미기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
plt.legend(loc='best', ncol=2, fontsize=16, frameon=True, shadow=True)
# frameon=False로 지정하면 테두리가 표시되지 않습니다.
# shadow 파라미터를 사용해서 텍스트 상자에 그림자를 표시할 수 있씁니다.

plt.show()
```
